```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

### Data: <a class = "anchor" id = "data"></a>

```python
Player = [
    "Rickey Henderson",
    "Lou Brock",
    "Ty Cobb",
    "Eddie Collins",
    "Max Carey",
    "Joe Morgan",
    "Luis Aparicio",
    "Paul Molitor",
    "Roberto Alomar"
]

SB = [1406, 938, 897, 741, 738, 689, 506, 504, 474]
CS = [335, 307, 212, 195, 109, 162, 136, 131, 114]
G = [3081, 2616, 3034, 2826, 2476, 2649, 2599, 2683, 2379]
```

```python
df = pd.DataFrame(data = zip(Player,G,SB,CS),
                  columns = ["Player", "G", "SB", "CS"])

# compute number of stolen base attempts
df["SB_Attempts"] = df["SB"] + df["CS"]

# compute SB success rate
df["SB_Rate"] = df["SB"] / df["SB_Attempts"]

# compute SB per game rate
df["SB_per_G"] = df["SB"] / df["G"]

df
```

```python
for i, r in enumerate(df):
    print(i,r)
```

```python
plt.figure(figsize = (12,8))

plt.scatter(x = df["SB_per_G"],
           y = df["SB_Rate"])

plt.title("Measuring Stolen Bases per Game against Stolen Base Success Rate")
plt.xlabel("Stolen Bases per Game")
plt.ylabel("Stolen Base Success Rate")

for i in df.index:
    plt.annotate(df["Player"][i], 
                 (df["SB_per_G"][i], df["SB_Rate"][i]));
```
